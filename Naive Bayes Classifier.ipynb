```python
import numpy as np
import csv
import math
```

```python
def data_loader(path):
  samples = csv.reader(open(path, 'rt'))
  samples = list(samples)

  feature = list()
  label = list()

  for sample in samples:
    sample = list(map(float, sample))
    label.append(int(sample.pop()))
    feature.append(sample)
  
  return feature, label
```

```python
def get_GaussianNBC(train_samples, train_labels):
  wine_class_0_samples = []
  wine_class_1_samples = []
  wine_class_2_samples = []

  for k in range(len(train_samples)):
    sample = train_samples[k]
    label = train_labels[k]

    if label == 0:
      wine_class_0_samples.append(sample)
    elif label == 1:
      wine_class_1_samples.append(sample)
    elif label == 2:
      wine_class_2_samples.append(sample)

  sample_by_classes = [
    wine_class_0_samples,
    wine_class_1_samples,
    wine_class_2_samples,
  ]
  print(np.array(sample_by_classes[0]).shape)
  num_classes = 3
  means_by_classes = []
  stds_by_classes = []

  for C in range(num_classes):
    means = []
    stds = []

    for features in zip(*sample_by_classes[C]):
      means.append(np.mean(features))
      stds.append(np.std(features))

    means_by_classes.append(means)
    stds_by_classes.append(stds)
  
  return means_by_classes, stds_by_classes
```

```python
def Gaussian_PDF(x, mean, std):
  return np.where(std == 0.0, np.where(x == mean, 1.0, 0.0), 1 / np.sqrt(2 * np.pi * np.power(std, 2)) * np.exp(-np.power(x - mean, 2) / (2 * np.power(std, 2))))
```

```python
def get_Acc(pred, label):
  acc = np.equal(pred, label)
  return list(acc).count(True) / len(acc) * 100
```

```python
train_set_path = '/content/drive/MyDrive/기계학습데이터/wine_train.csv'
test_set_path = '/content/drive/MyDrive/기계학습데이터/wine_test.csv'

train_samples, train_labels = data_loader(train_set_path)
test_samples, test_labels = data_loader(test_set_path)
means_by_classes, stds_by_classes = get_GaussianNBC(train_samples, train_labels)

pred_classes = predict(means_by_classes, stds_by_classes, test_samples)

acc = get_Acc(pred_classes, test_labels)
print('Acc:', acc)
```
